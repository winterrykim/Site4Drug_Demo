```python
from drugclip_af3_wrapper import build_pocket_lmdb_from_af3, run_drugclip_retrieval, read_ranked_compounds
import sys

rcsb = '4AG8'
start = '96'
end = '115'

work_dir = f"./runs/{rcsb}_{start}_{end}"
pocket_lmdb = f"{work_dir}/pocket.lmdb"

build_pocket_lmdb_from_af3(
    af3_dir=f"./{rcsb}",
    site_spec=f"{start}-{end}",
    pocket_lmdb_path=pocket_lmdb,
    pocket_name=f"{rcsb}_site_{start}_{end}",
    scratch_dir=f"{work_dir}/scratch",
)

ranked_path = run_drugclip_retrieval(
    drugclip_root="./DrugCLIP",
    checkpoint_path="../checkpoint_best.pt",
    pocket_lmdb_path=f"../runs/{rcsb}_{start}_{end}/pocket.lmdb",
    mols_lmdb_path="../mols-005.lmdb",
    emb_dir=f"../runs/{rcsb}_{start}_{end}/emb",
    data_dir="data",
    results_path=f"../runs/{rcsb}_{start}_{end}/results",
    batch_size=8,
    max_pocket_atoms=256,
    cuda_visible_devices="0",
    python_bin=sys.executable,
)

hits = read_ranked_compounds(ranked_path, smiles_to_name=None, top_k=20)
for x in hits[:10]:
    print(x)
```
